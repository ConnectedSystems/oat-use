```python
def V_i_regular(a):
    return (1./3.)/((1. + a)**2)
```

```python
def V_Ti_regular(V, i):
    result = 1.
    for j in range(len(V)):
        if j != i:
            result = result * (1 + V[j])
    return result * V[i]
```

```python
def V_regular(V):
    V = 1 + V
    result = np.prod(V) - 1
    return result
```

```python
def V_i_modified(alpha, a):
    return (alpha**2)/((1 + 2 * alpha) * (1 + a)**2)
```

```python
a = np.array((100, 0, 100, 100, 100, 100, 1, 0, 100, 100, 0, 100, 100, 100, 1, 100, 100, 0, 100, 1))
```

```python
alpha = np.array((1, 4, 1, 1, 1, 1, 0.5, 3, 1, 1, 2, 1, 1, 1, 0.5, 1, 1, 1.5, 1, 0.5))
```

```python
V = V_i_modified(alpha, a)
```

```python
V_total = V_regular(V)
```

```python
for i in range(20):
    print np.round(V_Ti_regular(V, i)/V_total, decimals = 4)
```
